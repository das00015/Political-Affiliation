```python
import json
```

```python
def get_hashtag_content(row):
    temp_list = json.loads(row)
    #print(temp_list)
    if len(temp_list)==0: 
        return ("empty",0,[])
    else:
        count_hashtags = len(temp_list)
        list_hashtags = []
        for every_dict in temp_list:
            #print(every_dict['text'])
            list_hashtags.append(every_dict['text'])
        return ("hashtags present",count_hashtags,list_hashtags)
    
    
```

```python
def replace_string(old,new,string):
    return string.replace(old,new)
```

```python
new_content = replace_string("'", '"',"[{'text': 'EndSARS', 'indices': [55, 63]}]")
print(new_content)
```

```python
get_hashtag_content(new_content)
```

```python
subdf['new_content'] = subdf.apply(lambda x : replace_string("'", '"',x['Hashtags']),axis=1)
```

```python
subdf['hashtag_content'] = subdf.apply(lambda x : get_hashtag_content(x['new_content']),axis=1)
```

```python
subdf['hashtag_content'].value_counts()
```

```python
hashtag_dict = {}

def get_hashtags(df):
    for index, row in df.iterrows():
        hashtags = row['hashtag_content'][2]
        for hasht in hashtags:
            if hasht in hashtag_dict:
                hashtag_dict[hasht]+=1
            else:
                hashtag_dict[hasht]=1
    print(hashtag_dict)
    
```

```python
get_hashtags(subdf)
```
