```python
#specify datatypes so that datatype guessing isnt done (looks like he has mentioned default returned value as 'none' string everywhere)
dtype={'UID': object, 'Datetime':object, 'User Name':object, 'Screen Name':object, 'Description':object, 'URLs':object,
       'Tweet Text':object, 'Tweet Lang':object, 'Trigrams':object, 'Profile Picture URL':object,
       'Original Text':object, 'No User Mentions':object, 'No URLs':object, 'No Tweets':object,
       'No Replies':object, 'No Likes':object, 'No Hashtags':object, 'No Friends':object, 'No Followers':object,
       'Mentions':object, 'Hashtags':object, 'Clean Tweet':object, 'Bigrams':object, 'TweetID':object,
       'TweetIDStr':object, 'truncated':object, 'Location':object, 'TweetReplyNo':object, 'TweetReTweetNo':object,
       'TweetFavoriteNo':object, 'RetweetedStatus':object}
```

```python
#Merge all the csvs in this folder together to create SED_vpdebate
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
print(all_filenames)
```

```python
#TEST : Lets read each one by one and check the number of columns 
from dask import dataframe as dd
for csv_name in all_filenames:
    ##Experiments with dask ##
    start = time.time()
    dask_df = dd.read_csv(csv_name)
    end = time.time()
    print("Read this csv with dask: ",(end-start),"sec")
    if len(dask_df.columns) !=31:
        print("err_df",csv_name)
        print(dask_df.columns)
        print("---")
        break
    
```

```python
#concat the data with pd
start = time.time()
combined_csv = pd.concat([pd.read_csv(f,dtype=dtype,lineterminator='\n') for f in all_filenames])
end = time.time()
print("Concat csv with pandas: ",(end-start),"sec")
```
