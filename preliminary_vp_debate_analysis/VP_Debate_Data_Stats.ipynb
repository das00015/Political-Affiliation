```python
def replace_string(old,new,string):
    return string.replace(old,new)
```

```python
def get_mention_content(row):
    #print(row)
    try:
        temp_list = json.loads(row)
        #this will work for all cases, except in the case of a name such as "Jason Mc'Neely" which would have turned into "Jason Mc" Neely" because of replace. 
        if len(temp_list)==0: 
            return (0,[])
        else:
            count_mentions = len(temp_list)
            list_mentions = []
            for every_dict in temp_list:
                list_mentions.append(every_dict['name'])
            return (count_mentions,list_mentions)
    except:
        return "Error Case"
    
```

```python
mendf['new_content'] = mendf.apply(lambda x : replace_string("'",'"',x['Mentions']),axis=1)
```

```python
def get_hashtag_content(row):
    temp_list = json.loads(row)
    #print(temp_list)
    if len(temp_list)==0: 
        return ("empty",0,[])
    else:
        count_hashtags = len(temp_list)
        list_hashtags = []
        for every_dict in temp_list:
            #print(every_dict['text'])
            list_hashtags.append(every_dict['text'])
        return ("hashtags present",count_hashtags,list_hashtags)
    
    
```

```python
def get_url_content(row):
    try:
        temp_list = json.loads(row)
        if len(temp_list)==0: 
            return (0,[])
        else:
            count_urls = len(temp_list)
            list_urls = []
            for every_dict in temp_list:
                list_urls.append(every_dict['url'])
            return (count_urls,list_urls)
    except:
        return "Error Case"
```

```python
udf['new_content'] = udf.apply(lambda x : replace_string("'",'"',x['URLs']),axis=1)
```

```python
udf['url_content'] = udf.apply(lambda x : get_url_content(x['new_content']),axis=1)
```
