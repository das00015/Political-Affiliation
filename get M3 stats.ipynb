```python
import csv
import pandas as pd
import json
import ast
import operator
```

```python
input_filename = "testoutput_100_mod.csv"
```

```python
df = pd.read_csv (input_filename)
df.head()
```

```python
df.dtypes
```

```python
def get_age_stats(row):
    temp_dict = ast.literal_eval(row)
    max_assigned_age = max(temp_dict.items(), key=operator.itemgetter(1))[0]
    return max_assigned_age
    
    
```

```python
def get_gender_stats(row):
    temp_dict = ast.literal_eval(row)
    max_assigned_gender = max(temp_dict.items(), key=operator.itemgetter(1))[0]
    return max_assigned_gender
    
```

```python
def get_org_stats(row):
    temp_dict = ast.literal_eval(row)
    max_assigned_org = max(temp_dict.items(), key=operator.itemgetter(1))[0]
    return max_assigned_org
    
```

```python
df['age_assigned'] = df['age'].apply(get_age_stats)
```

```python
df['org_assigned'] = df['org'].apply(get_org_stats)
```

```python
df['gender_assigned'] = df['gender'].apply(get_gender_stats)
```
